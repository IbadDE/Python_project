```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
url = 'https://stats.espncricinfo.com/ci/engine/records/team/match_results.html?id=14395;type=tournament'
source = requests.get(url)
soup = bs(source.text, 'html.parser')

matches_21_22 = soup.find('tbody').find_all('tr', class_ = 'data1')
print(len(matches_21_22))


match_21_22 = {
    'Team_1' : [],
    'Team_2' : [],
    'Winner' : [],
    'Margin' : [],
    'Ground' : [],
    'Match_date' : [],
}


for row in matches_21_22:
    all_td = row.find_all('td', recursive=False)
    

    match_21_22['Team_1'].append(all_td[0].text)
    match_21_22['Team_2'].append(all_td[1].text)
    match_21_22['Winner'].append(all_td[2].text)
    match_21_22['Margin'].append(all_td[3].text)
    match_21_22['Ground'].append(all_td[4].text)
    match_21_22['Match_date'].append(all_td[5].text)
    
    


df_match_21_22 = pd.DataFrame(match_21_22)
df_match_21_22["Season"] = "2021/2022"
print(df_match_21_22.shape)
df_match_21_22.head()
```
