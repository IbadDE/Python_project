```python
import pandas as pd
from bs4 import BeautifulSoup as bs
import requests
```

```python
url = 'https://www.espncricinfo.com/series/pakistan-super-league-2021-22-1292999/points-table-standings'
source = requests.get(url)
soup = bs(source.text, 'html.parser')
soup

tbody_tag = soup.find('tbody')
tr_tag = tbody_tag.find_all('tr',class_ = "ds-text-tight-s")

tbl_21_22 = {
    'team' : [],
    'matches' : [],
    'win': [],
    'loss' : [],
    'tie' : [],
    'no_result' : [],
    'points' : [],
    'nrr' : [],
    'for' : [],
    'against' : []
}


for row in tr_tag:
    td_tag = row.find_all('td', recursive=False)
    tbl_21_22['team'].append(td_tag[0].text)
    tbl_21_22['matches'].append(td_tag[1].text)
    tbl_21_22['win'].append(td_tag[2].text)
    tbl_21_22['loss'].append(td_tag[3].text)
    tbl_21_22['tie'].append(td_tag[4].text)
    tbl_21_22['no_result'].append(td_tag[5].text)
    tbl_21_22['points'].append(td_tag[6].text)
    tbl_21_22['nrr'].append(td_tag[7].text)
    tbl_21_22['for'].append(td_tag[9].text)
    tbl_21_22['against'].append(td_tag[10].text)

df_22 = pd.DataFrame(tbl_21_22)

df_22['season'] = "2021/2022"
df_22
```
